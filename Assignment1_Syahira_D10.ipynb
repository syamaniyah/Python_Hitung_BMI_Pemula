```python
#2.	Hitunglah BMI setiap siswa dengan rumus BMI=(Berat badan (kg))/((Tinggi Badan (m))^2 )
#3.	Buatlah program untuk menghitung BMI!
def hitung_bmi(row):
    berat_badan = row['Berat Badan (Kg)']
    tinggi_badan = row['Tinggi Badan (Cm)'] / 100.0

    bmi = berat_badan / (tinggi_badan**2)
    return bmi
```

```python
df_d['BMI'] = df_d.apply(lambda row: hitung_bmi(row), axis = 1)
```

```python
#4.	Update dataframe df dengan menambahkan hasil perhitungan soal nomor 2, beri nama kolom BMI.
df_d
```

```python
#5.	Buatlah status BMI untuk setiap siswa dari kategori berikut, lalu simpan data statusnya dengan nama kolom Status BMI.
#	Normal: 18.5-24.9
#	Overweight: 25.0-29.9
#	Obesity: 30.0-34.9
#	Underweight: < 18.5

def Status_BMI(BMI):
    #if bmi >= 18.5 and bmi <= 24.9
    if BMI<18.5:
        return'Underweight'
    elif 18.5<=BMI<=24.9:
        return'Normal'
    elif 25.0<=BMI<=29.9:
        return 'Overweight'
    elif 30.0<=BMI<=34.9:
        return'Obesity'
```

```python
df_d['Status_BMI']=df_d['BMI'].apply(Status_BMI)
```
